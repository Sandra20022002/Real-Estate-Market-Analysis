# Исследование объявлений о продаже квартир

В нашем распоряжении есть данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно определить рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

Возможные причины пропусков:

В столбцах 'is_apartment', 'balcony' причины пропусков могли быть следующими:
- техническая ошибка  
- данные не считались важными для продажи 
- у заполнявшего не было нужной информации (например, пользователь не понял, что значит 'аппартаменты') 
- если значение было нулевым (то есть в квартире нет балконов, квартира не является аппартаментом)

В столбцах living areа, kitchen_area, ceiling_height  значения могли быть пропущены по следующим причинам:
- техническая ошибка
- пользователь не захотел или по невнимательности забыл внести данные
- у пользователя не было нужной информации

В столбцах 'living_area', 'kitchen_area' заменим пропуски c помощью двух медиан:
- медианы отношения общих площадей к жилой площади 
- медианы отношения общих площадей к площади кухни

Рассчитаем медианы на базе известных значений:

```python
ratio_total_living = data['total_area']/data['living_area']
ratio_total_kitchen = data['total_area']/data['kitchen_area']
print (ratio_total_living.median())
print (ratio_total_kitchen.median())
```

До замены приведем все отрицательные числа к положительным (abs() и заменим ошибочные значения на NaN (to_numeric). Затем используем функцию fillna () и подставим медиану соотношений.

```python
data['ceiling_height'] = data['ceiling_height'].abs()
data['ceiling_height'] = data['ceiling_height'].apply(pd.to_numeric, errors='coerce')

median = data.loc[data['ceiling_height'] <= 10, 'ceiling_height'].median()
data.loc[data['ceiling_height'] > 10, 'ceiling_height'] = np.nan
data['ceiling_height'].fillna(median,inplace=True)
```

Создадим функцию, которая сортирует этажность квартир: первый, последний, другой. 

```python
def floor_descr(row):
    try:    
        if row['floor'] == 1:
            return 'Первый'
        if row['floor']  == row['floors_total']:
            return 'Последний'
        else:
            return 'Другой'
    
    except:
        return 'Нет информации'

data['floor_description'] = data.apply (lambda row: floor_descr(row), axis=1)   
```
